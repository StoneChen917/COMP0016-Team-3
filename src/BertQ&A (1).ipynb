```python
from transformers import BertForQuestionAnswering
model = BertForQuestionAnswering.from_pretrained('deepset/bert-base-cased-squad2')
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('deepset/bert-base-cased-squad2')
```

```python
questions = ["What is the Operation Start Date?", "What is the Operation End Date?", "What is the Glide Number?", "What is the Operation n°?", "What is the Operation Budget?", "What is the Country of Disaster?", "What is the Host National Society?"]

#questions
#questions1 = "What is the Operation Start Date?"
#questions2 = "What is the Operation End Date?"
#questions3 = "What is the Glide Number?"
#questions4 = "What is the Operation n°?"
#questions5 = "What is the Operation Budget?"
#questions6 = "What is the Country of Disaster?"
#questions7 = "What is the Host National Society?"
# number of people affectes
# number of people assisted
```

```python
tokenizer.encode(questions[0], truncation = True, padding = True)
```

```python
tokenizer.encode('[CLS]')
```

```python
from transformers import pipeline
nlp = pipeline('question-answering', model = model, tokenizer = tokenizer)
#nlp = pipeline("question-answering", model='distilbert-base-cased-distilled-squad')
```

```python
# importing required modules
from PyPDF2 import PdfReader

# creating a pdf file object
pdfFileObj = open('sample1.pdf', 'rb')

# creating a pdf reader object
pdfReader = PdfReader(pdfFileObj)

# printing number of pages in pdf file
print(len(pdfReader.pages))

# creating a page object
pageObj = pdfReader.pages[0]

# extracting text from page
context = pageObj.extract_text()


# closing the pdf file object
pdfFileObj.close()

print(context)
```

# New Section

```python
nlp({'question': questions[0], 'context': context})
```
